```python
# dataframe 정렬하기
import pandas as pd
data = pd.DataFrame({"cluster" : [1,1,2,1,2,3], "org":['a','a','h','c','d','w'], "time":[8,6,34,23, 74,6]})
```

```python
# time을 기준으로 내림차순으로 정렬한다.
data.sort_values(['time'], ascending=False)
```

```python
# 두개 열을 기준으로 정렬할 수도 있다.
data.sort_values(['cluster', 'org'], ascending=True)
```

```python
# groupby 다루기
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'city': ['부산', '부산', '부산', '부산', '서울', '서울', '서울'],
    'fruits': ['apple', 'orange', 'banana', 'banana', 'apple', 'apple', 'banana'],
    'price': [100, 200, 250, 300, 150, 200, 400],
    'quantity': [1, 2, 3, 4, 5, 6, 7]
})
df
```

```python
# 해당 열을 기준으로 묶어 마지막 열을 계산한다.
df.groupby('fruits').mean()
```

```python
df.groupby('city').mean()
```

```python
df.groupby(['city', 'fruits']).mean()
```

```python
# groupby를 사용하면 기본으로 그룹 라벨이 index가 됩니다.
# index를 사용하고 싶은 않은 경우에는 as_index=False 를 설정하면 됩니다.
df.groupby(['city', 'fruits'], as_index=False).mean()
```

```python
# 그룹 안에 데이터를 확인하고 싶은 경우 사용한다.
df.groupby('city').get_group('부산')
```

```python
# 그룹의 크기 확인
df.groupby('city').size()
```

```python
# aggregation을 이용하여 계산할 수 있다.
df.groupby('city').agg(np.mean)
```

```python
# 한 열은 평균을 한 열은 합계를 나타내고 싶을 때 사용
df.groupby('city').agg({'price': np.mean, 'quantity': np.sum})
```

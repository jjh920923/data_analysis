```python
# 다차원 척도법(MDS : Multi Dimensional Scaling)
# 차원의 축소를 통해 개체들의 상대적 위치 등을 통해 개체들 사이의 관계를 쉽게 파악하고자 하는데 목적이 있다.

# 계량적 MDS(Metric MDS)
# 데이터가 등간척도나 비율척도인 경우 활용
# 각 개체들간의 유클리드 거리행렬을 계산

# 비계량적 MDS(Non-Metric MDS)
# 데이터가 순서척도인 경우 활용
# 순서척도를 거리의 속성처럼 변환해서 분석
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import mds
from scipy.spatial.distance import pdist, squareform
import PIL.Image as pilimg
import warnings
warnings.simplefilter("ignore")
```

```python
arrest = pd.read_csv('USArrests.csv')
```

```python
arrest = arrest.set_index('Unnamed: 0')
```

```python
# 결측치 확인
arrest.isnull().any()
```

```python
# 데이터 타입 확인
arrest.dtypes
```

```python
# 데이터 크기 확인
np.shape(arrest)
```

```python
# boxplot을 이용한 이상치 검출
arrest.boxplot()
plt.show()
```

```python
# 데이터 범위를 맞춰주기 위한 스케일링 작업
scaler = StandardScaler()
scaled = scaler.fit_transform(arrest)
df = pd.DataFrame(scaled)
df.columns = arrest.columns
df.index = arrest.index
df.head()
```

```python
# 스케일링 작업 후 이상치 확인
# Rape 변수에 두가지 이상치가 확인되지만 극단적인 이상치가 아니기 때문에 제거하지않는다.
df.boxplot()
plt.show()
```

```python
# 스트레스 지수를 이용하여 아래와 같은 코드로 적절한 MDS 차원을 정할 수 있다.
# 이 데이터는 변수의 갯수가 적어 적절한 component의 수가 4개지만 연습을 위해 2개로 줄이는 방법을 택하겠다.
stress = []
for i in range(1, 11):
    stress.append(mds.MDS(n_components=i).fit(df).stress_)
plt.plot(range(1, 11), stress)
plt.xlabel('k')
plt.ylabel('stress : residual sum of squares')
print('optimal numbers of components: ', np.argmin(stress)+1)
```

```python
# 다차원 척도법을 이용한 변수 축소
model = mds.MDS().fit(df)
df_mds = model.embedding_
df_mds = pd.DataFrame(df_mds)
df_mds.index = arrest.index
df_mds.head()
```

```python
# MDS를 이용하여 2차원으로 줄이면 데이터들의 관계를 그래프로 한눈에 파악할 수 있다.
x, y = df_mds[0], df_mds[1]
plt.figure(figsize=(8, 6))
plt.scatter(x, y)
plt.xlabel('component1')
plt.ylabel('component2')
plt.title('Metric MDS')
for i in range(df_mds.shape[0]):
    plt.text(df_mds[0][i], df_mds[1][i], df_mds.index[i])
plt.show()
```

```python
# 위의 표를 통해 스트레스 지수에 따른 적합도 검정을 해보겠다.
# 아래는 데이터를 거리행렬로 변환 후 스트레스 지수를 구하는 공식이다.
# 출력된 배열의 첫번째는 component가 하나일 때 28%의 스트레스 지수가 나타난 것이고
# 두번째는 component가 두개일 때 0.3%의 스트레스 지수가 나타난 것이다.
# component가 두개일 때 완적히 적합된 것을 확인할 수 있다.

dist = pdist(df_mds, 'euclidean')
df_dist = pd.DataFrame(squareform(dist))
d = np.square(df_dist).sum().sum()/2
stress_std = np.sqrt(stress/d)
stress_std
```

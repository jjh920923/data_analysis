```python
# 이상치(outlier)는 데이터의 분포에서 비정상적으로 분포를 벗어난 값이다.

# 이상치가 발생하는 이유에는 4가지 정도가 있다.
# 1. 의도하지 않게 잘못 입력한 경우
# 2. 의도하지 않게 입력됐으나 분석 목적에 부합되지 않아 제거해야 하는 경우
# 3. 의도되지 않은 현상이지만 분석에 포함해야 하는 경우
# 4. 의도된 이상값인 경우(대부분 사기(fraud)이다.)

# 이상치를 탐지하는 방법에는 여러가지가 있다.
# 그 중에서 
# 1. boxplot을 이용하여 검출한다.
# 2. 정규분포에서 일반적으로 +, - 3sigma에 해당하는 데이터를 이상치로 판단한다.
# 3. 베이즈 정리에 의해 데이터 셋이 가지는 두가지 샘플(정상/이상)에 대한 
#    발생 확률(likelihood)로 이상치 판별
# 4. Nearest-neighbor : 모든 데이터 쌍의 거리를 계산하여 이상치 검출
#    다변수 분석일 경우 PCA or MDS를 활용하여 2차원으로 축소한 뒤 시행한다.
# 5. 샘플의 LOF(local outlier factor)를 계산하여 값이 가장 큰 데이터를 이상치로
#    추정(밀도기반으로 검출하는 방식)

# 이상치를 대체하는 방법
# 1. 일반적으로 +, - 3sigma에 해당하는 값을 이상치로 판단하여 제거하거나 대체한다.
#    하한값 = 평균 - n*표준편차, 상한값 = 평균 + n*표준편차
# 2. 중위수로 부터 n편차 큰 값을 대체
# 3. 상위 p번째 백분위수보다 큰 값을 대체

# 이상치를 제거할 때 너무 많은 데이터가 손실된다면 모델 생성에 영향을 미친다.
# 그렇기 때문에 의미있는 이상치인지 의미없는 이상치인지 비즈니스 관점으로 판단을
# 한 후에 제거 or 대체 할 지 판단해야 한다.

# 참고 : https://pubdata.tistory.com/52
```

```python
import numpy as np
import pandas as pd
import statistics
import PIL.Image as pilimg
```

```python
# 임의의 정규분포 데이터 100개 생성
data = {"a" : np.random.normal(size = 100), "b" : np.random.normal(size = 100)} 
df = pd.DataFrame(data)
df
```

```python
# boxplot을 이용한 이상치 제거
df.boxplot()
```

```python
def boxplot_outlier(df):
    for i in range(np.shape(df)[1]):     
        Q1 = df.iloc[:,i].quantile(0.25)
        Q3 = df.iloc[:,i].quantile(0.75)
        IQR = Q3 - Q1
        outlier_left = float(Q1 - (1.5 * IQR))
        outlier_right = float(Q3 + (1.5 * IQR))
        outlier = df[(df.iloc[:,i] < outlier_left) | (df.iloc[:,i] > outlier_right)]
        df.iloc[:,i][outlier.index] = np.nan
    df_new = df.dropna()
    return df_new
```

```python
df_outlier = boxplot_outlier(df)
np.shape(df_outlier)
```

```python
# 정규분포를 통한 이상치 제거
# 일반적으로 3sigma 기준으로 이상치를 검출한다.
def sigma3_outlier(df):
    for i in range(np.shape(df)[1]):     
        std = statistics.stdev(df.iloc[:,i])
        sigma3_left = -3*std
        sigma3_right = 3*std
        outlier = df[(df.iloc[:,i] < sigma3_left) | (df.iloc[:,i] > sigma3_right)]
        df.iloc[:,i][outlier.index] = np.nan
    df_new = df.dropna()
    return df_new
```

```python
df_outlier = sigma3_outlier(df)
np.shape(df_outlier)
```

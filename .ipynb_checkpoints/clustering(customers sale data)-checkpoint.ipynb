```python
import pandas as pd
import scikitplot as skplt
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Wholesale customers data는 지역별 생필품 소비량을 나타낸 데이터이다.
# 결측치 확인 결과 없는 것으로 나타났다.
df = pd.read_csv("Wholesale customers data.csv")
df.isnull().sum()
```

```python
# 지역별로 생필품을 소모하는 양이 다르기 때문에 차이가 나는 것으로 보인다.
# 군집분석에서 이상치는 우리가 이전에 공부했던 이상치와는 기준이 조금 다르다.
# 이전의 이상치는 일정 범위에서 벗어난 값을 이상치라고 불렀지만 군집분석에서의 이상치는
# 군집에서 벗어난 데이터를 이상치로 취급한다.
df.boxplot()
```

```python
categorical_features = ['Channel', 'Region']
continuous_features = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']
for col in categorical_features:
    dummies = pd.get_dummies(df[col], prefix=col)
    df = pd.concat([df, dummies], axis=1)
    df.drop(col, axis=1, inplace=True)
df
```

```python
# 변수간의 범위가 다르기 때문에 스케일링을 해준다.
# 더미변수로 바꿔줬기 때문에 더미변수의 데이터 유지를 위해
# MinMaxScaler를 해주도록 한다.
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df)
scaled = pd.DataFrame(scaled)
scaled
```

```python
# 먼저 군집의 수를 정하기 위한 척도를 구해본다.
# 그래프를 확인해보면 5~6개로 예상이되는데 정확한 수치를 확인해보기 위해서
# 실루엣을 통해 알아보자.
Sum_of_squared_distances = []
K = range(1, 15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(scaled)
    Sum_of_squared_distances.append(km.inertia_)
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
silhouette = []
k_range = range(2, 10)
for k in k_range:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(scaled)
    silhouette_avg = silhouette_score(scaled, kmeans.labels_)
    silhouette.append(silhouette_avg)
silhouette
```

```python
# 그래프로 확인해 본 결과 군집의 수가 6개 일 때 가장 높은 것으로 나타났다.
plt.plot([1, 2], [0, silhouette[0]], c='blue')
plt.plot(k_range, silhouette, 'bx-')
plt.xlim(1, 9)
plt.xlabel('Number of cluster k')
plt.ylim(0, 1)
plt.ylabel('average silhouette width')
plt.title('Optibal number of clusters')
```

```python
# kmeans를 사용하여 군집의 수를 6개로 잡고 분석한 결과이다.
# 각 중심과 어느 군집에 속하는지 알 수 있다.
km = KMeans(n_clusters=6)
km.fit(scaled)
print(km.cluster_centers_)
print(km.labels_)
```

```python
# 실루엣 스코어가 약 0.85로 아주 잘 분류된 것을 확인할 수 있다.
# 각 군집마다 50%이상이 빨간점선 안에 속하면 좋은 모델이라 할 수 있겠다.
pred = km.fit_predict(scaled)
skplt.metrics.plot_silhouette(scaled, pred)
```

```python
# DBSCAN을 이용한 분석 결과
db = DBSCAN()
db.fit(scaled)
db.labels_
```

```python
# 계층적 군집분석을 통한 결과
AC = AgglomerativeClustering(n_clusters=6, affinity='euclidean', linkage='ward')
AC.fit_predict(scaled)
```

```python
# 계층적 군집방법을 통해 덴드로그램을 그려보았다. 행의 수가 많아 잘 확인되지는 않지만
# 6개의 군으로 분류되는 것은 명확하게 볼 수 있다.
linked = linkage(scaled, 'ward')
labelList = range(1, 441)
plt.figure(figsize=(30, 21))
dendrogram(linked, orientation='top', labels=labelList, 
           distance_sort='descending', show_leaf_counts=True)
```
